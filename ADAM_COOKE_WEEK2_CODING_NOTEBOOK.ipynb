What happens when you don't have `position = "identity"` in your `geom_histogram` layer? Answer below in a `Markdown` cell.

___
When you do not specifically indicate **position = "*identity*"** in the geom_histogram function, ggplot will use the default value of **position = "*stack*"**. Stack will "stack" the bars over each other, while Identity will layer the bars on top of each other with the option of transparency when specifying a value for alpha. 


**position = "*stack*"** (default) is more suited for visualizing total counts and overal distribution. 

**position = "*identity*"** is better for visualizing multiple distributions to see the overlapping of different groups of data.

## Data wrangling

First we are going to bring in a data set to practice with called `txhousing`. You can find some information about `txhousing` [here](https://ggplot2.tidyverse.org/reference/txhousing.html).

```r
# uncomment the next two lines and execute this cell
data(txhousing)
glimpse(txhousing)
```

Write a code chunk to remove the inventory variable. Save the results in a data frame called `txhousing`. Use `glimpse` or `head` to confirm you have removed the variable.

```r
txhousing <- txhousing |> select(-"inventory")

glimpse(txhousing)
```

Make a data set called `dallas_sub` that includes data only from the city of Dallas in 2012 & 2013.

```r
dallas_sub <- txhousing |> filter(city == "Dallas", year %in% c(2012, 2013))
glimpse(dallas_sub)
```

Add a column to the `dallas_sub` data set called `prct_sold` that calculates the percentage of listings that were sold (`sales/listings * 100`). Be sure to save the results also as a data frame called `dallas_sub`.

```r
dallas_sub <- dallas_sub |> mutate(prct_sold = (sales / listings * 100))
glimpse(dallas_sub)
```

Calculate the average percentage of listings that were sold in Dallas in each month of the year based on your `dallas_sub` data set. Save the results of the calculation in an data frame called `dallas_summary`.

```r
dallas_summary <- dallas_sub |> group_by(month) |> summarize(avg_pct_sold_per_month = mean(prct_sold))
dallas_summary
```

Use `slice_max` to see the top 3 months with the greatest average percentage of listings sold from `dallas_summary`. You do not need to save the results.

```r
dallas_summary <- dallas_summary |> slice_max(avg_pct_sold_per_month, n = 3)
dallas_summary
```

Run the following code chunk. Study the code, and the output. Explain in your own words what this code chunk calculated.

```r
 txhousing |>
   filter(year == 2012 | year == 2013, city == "Dallas") |>
   mutate(prct_sold = sales/listings *100) |>
   group_by(month) |>
   summarize(mean_prct_sold = mean(prct_sold)) |>
   slice_max(mean_prct_sold, n = 3)
##txhousing
```

___
This code cell is essentially doing everything that we have done in the previous cells all in one cell. 

Bringing it all together, this code is **FILTERING** the *txhousing* data based where the Year values are *equal* to 2012 **OR** 2013, **AND** the City values are equal to "Dallas". From that filtered data, this code is adding a new column using the **MUTATE** method, that calculates the percentages of houses sold according to how many are listed. With that new column created, the dataset is then grouped by month using the **GROUP_BY** method, and the average percentage of houses sold is calculated with the **summarize** method. After all of that is said and done, the data is filtered again using the **SLICE_MAX** method to select the three months with the highest percentages of houses sold. Ultimately, this code is only going to return 3 rows of records, and only show the *month* and *mean_prct_sold* columns because that is what is being grouped by and calculated.
